```python
import pandas as pd


jobs = pd.read_csv("JosephDAbishak_JOY_UNIVERSITY_Jobs_Dataset.csv")

print("Total jobs loaded:", len(jobs))
jobs.head()
```

```python
jobs["text"] = jobs["job_title"] + " " + jobs["skills"]
jobs.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(stop_words="english")


job_vectors = vectorizer.fit_transform(jobs["text"])

print(" NLP model trained")
print("Vector shape:", job_vectors.shape)
```

```python
user_input = input("Enter your skills (comma separated): ")
```

```python
from sklearn.metrics.pairwise import cosine_similarity


user_vector = vectorizer.transform([user_input])

similarity_scores = cosine_similarity(user_vector, job_vectors)


jobs["similarity"] = similarity_scores[0]


top_jobs = jobs.sort_values(by="similarity", ascending=False).head(5)

print("\n Top Job Recommendations:")
for i, row in top_jobs.iterrows():
    print("-", row["job_title"])
```

```python
from rapidfuzz import process
from sklearn.metrics.pairwise import cosine_similarity
```

```python

all_skills = (
    jobs["skills"]
    .str.lower()
    .str.split(", ")
    .explode()
    .unique()
)
```

```python
user_input = input("Enter your skills (comma separated): ").lower()

final_skills = []

for skill in user_input.split(","):
    skill = skill.strip()

    match = process.extractOne(skill, all_skills)

    if match and match[1] >= 80:

        final_skills.append(match[0])
    else:

        print(f"âš  Skill '{skill}' not clearly recognized.")
        clarify = input(
            f"Can you briefly explain or rephrase '{skill}'? (press Enter to skip): "
        )
        if clarify.strip():
            final_skills.append(clarify.strip())
        else:
            final_skills.append(skill)

cleaned_input = " ".join(final_skills)

print("\n Final processed skills:")
print(cleaned_input)
```

```python

domain_map = {
    "NLP": ["nlp", "text", "language", "chatbot"],
    "Computer Vision": ["vision", "image", "opencv", "cnn"],
    "Machine Learning": ["ml", "machine learning", "model"],
    "Data Analytics": ["sql", "excel", "analysis", "visualization"],
    "AI": ["ai", "artificial intelligence", "deep learning"]
}

detected_domains = set()

for domain, keywords in domain_map.items():
    for word in keywords:
        if word in cleaned_input:
            detected_domains.add(domain)

if detected_domains:
    print(" Detected domain(s):", ", ".join(detected_domains))
else:
    print(" Detected domain(s): General AI / Data Science")



user_vector = vectorizer.transform([cleaned_input])
similarity_scores = cosine_similarity(user_vector, job_vectors)

jobs["similarity"] = similarity_scores[0]
jobs["match_percent"] = (jobs["similarity"] * 100).round(2)



experience = input(
    "Enter experience level (Intern / Junior / Senior / Any): "
).lower()

if experience != "any":
    jobs_filtered = jobs[jobs["job_title"].str.lower().str.contains(experience)]
    if jobs_filtered.empty:
        jobs_filtered = jobs.copy()
else:
    jobs_filtered = jobs.copy()



try:
    top_n = int(input("How many job recommendations do you want? (3 / 5 / 10): "))
except:
    top_n = 5



top_jobs = jobs_filtered.sort_values(
    by="similarity", ascending=False
).head(top_n)



with open("new_skills_log.txt", "a") as f:
    for skill in final_skills:
        if skill not in all_skills:
            f.write(skill + "\n")



job_descriptions = {
    "AI Engineer": "Builds and deploys AI models for real-world applications.",
    "Machine Learning Engineer": "Develops and optimizes machine learning models for production.",
    "Data Scientist": "Analyzes data and builds predictive models for decision-making.",
    "NLP Engineer": "Works on text processing, language models, and chatbots.",
    "Computer Vision Engineer": "Develops systems that analyze images and videos."
}



print("\n Top Job Recommendations:\n")

user_skill_set = set(cleaned_input.split())

for _, row in top_jobs.iterrows():
    job_skill_set = set(row["skills"].lower().split(", "))
    matched = user_skill_set.intersection(job_skill_set)
    missing_skills = job_skill_set - user_skill_set

    description = job_descriptions.get(
        row["job_title"],
        "This role focuses on AI and data-driven problem solving."
    )

    suggested = list(missing_skills)[:3]

    print(f" {row['job_title']}")
    print(f"   Match: {row['match_percent']}%")
    print(f"   About this role: {description}")
    print(f"   Matched skills: {', '.join(matched) if matched else 'General relevance'}")
    print(
        f"   Suggested skills to improve: "
        f"{', '.join(suggested) if suggested else 'You already match well'}\n"
    )
```
